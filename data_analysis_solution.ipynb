```python
#import libraries:
import pandas as pd
```

### data gathring:

1. Read in the dataset and take a look at the top few rows here:

```python
#read the csv file 
df = pd.read_csv('tmdb-movies.csv')
df.head()
```

### data assessing:

b. Find the number of rows and columns in the dataset.

```python
#general infomration ,describe,info,shape
df.shape[0]
df.info()
df.describe()
```

```python
#drop unwanted columns: 
#[cast,overview,tagline,keywords,production_companies,homepage,imdb_id]
df.drop(['cast','overview','tagline','keywords','production_companies','homepage','imdb_id'], axis=1, inplace=True)
```

```python
#correct the wrong data type to a propper type
df['release_date'] = pd.to_datetime(df['release_date'])
df['budget'] = df['budget'].astype(int)
df['id'] = df['id'].astype(str)
df['popularity'] = df['popularity'].astype(float)
```

```python
#to print out a spesfic time
df.release_date.dt.weekday_name #.time/ dt.hour
```

f. How many missing values in the dataset?

```python
#null values
df.isnull().sum()
```

```python
#drop null values
df.dropna(inplace=True)
df.isnull().sum()
```

f. how many diplucated rows in the dataset?

```python
#diplicated values
df.duplicated().sum()
#to print out the duplicated rows:
df[df.duplicated(keep=False)]

```

```python
#drop diplicated rows
df.drop_duplicates(inplace= True)
df.duplicated().sum()
```

```python
#splitting genres column:
df['genres']=df['genres'].str.split('|', 1,expand=True)
```

```python
#most genres movies:
df.genres.value_counts()
```

```python
#visulaiztion: hist scatter,bar
df.genres.value_counts().plot.bar();
```

```python
df.budget.plot.hist();
```

```python
df.plot.scatter("budget","revenue" );
```

```python
#filter/mask/query: you can do it by many methods:
#1
df2=df[df['genres']=='Action']
#2
df2=df.query('genres=="Action"')
df2
```

```python
#grouping:
df3=df.groupby('genres')['popularity','revenue'].mean()

df3
```

```python
#export the dataframe to csv file
df.to_csv('datamovies.csv' , index=False)
```
